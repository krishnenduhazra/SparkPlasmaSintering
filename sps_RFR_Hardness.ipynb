### Import Dependency

```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
```

### Data Import

```python
df=pd.read_csv('sps3.csv')
print(df)
df.columns
```

```python
X=df.drop(['Hardness', 'FT','RH'],axis=1).values
y=df['Hardness'].values
```

### Scaling and Splitting

```python
scaler = StandardScaler()
scaler.fit(X)  
X = scaler.transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Model training

```python
regr = RandomForestRegressor(n_estimators=18, max_depth=10, random_state=0)
regr.fit(X_train,y_train)
y_pred=regr.predict(X_test)
```

### Accuracy, MSE & MAE

```python
accuracy=r2_score(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(accuracy,mse,mae)
```

### Feature Importance

```python
feature_list = list(['Temperature', 'RD', 'Grain_size'])
feature_imp = pd.Series(regr.feature_importances_, index=feature_list).sort_values(ascending=False)
print(feature_imp)
```

### Plots

```python
plt.figure(figsize=(10,6))
plt.scatter( y_test, y_pred)
plt.xlabel('Practical')
plt.ylabel('Predction')
plt.title('Hardness Practical Vs Predction')
```

```python
#Temperature Vs Hardness
temp=[row[0] for row in X_test]
dp={}
dt={}
for i in range(len(temp)):
    dp[temp[i]]=y_pred[i]
    dt[temp[i]]=y_test[i]

otemp=sorted(temp)
oy_pred=[]
oy_test=[]
for k in otemp:
    oy_pred.append(dp[k])
    oy_test.append(dt[k])
    
#Plot
plt.figure(figsize=(12,6))
plt.plot(otemp,oy_pred)
plt.scatter(otemp,oy_pred)
plt.plot(otemp,oy_test)
plt.scatter(otemp,oy_test)
plt.xlabel('Temperature')
plt.ylabel('Hardness')
plt.title('Temperature Vs Hardness')
plt.show()
```
